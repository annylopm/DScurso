Reto 2: Indexación de Series

```python
gastos_mensuales = {
    'A': 15000,
    'B': 200000,
    'C': 3250000,
    'D': 120000,
    'E': 135000,
    'F': 55000,
    'G': 100000,
    'H': 25000
}
```

```python
gastos_series = pd.Series(gastos_mensuales)
```

```python
# Los gastos de la división 'D' y 'G'
gastos_D_G = gastos_series.loc[['D', 'G']]
gastos_D_G
```

```python
# Los gastos de la división 'A' y 'E'
gastos_A_E = gastos_series.loc[['A', 'E']]
gastos_A_E
```

```python
# Los gastos de la división 'B', 'F' y 'H'
gastos_B_F_H = gastos_series.loc[['B', 'F', 'H']]
gastos_B_F_H
```

```python
# Los gastos desde la primera división hasta la división 'E'
gastos_principio_a_E = gastos_series.loc[:'E']
gastos_principio_a_E
```

```python
# Los gastos desde la división 'D' hasta la 'G'
gastos_D_a_G = gastos_series.loc['D':'G']
gastos_D_a_G
```

```python
# Los gastos desde la división 'C' hasta el la última división
gastos_C_a_final = gastos_series.loc['C':]
gastos_C_a_final
```
