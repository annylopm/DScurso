Reto 2: Automatizando con for loops

```python
def potencia_de_dos(exponente):
  resultado = 2

  if exponente == 0:
    return 1

  for i in range (0, exponente - 1):
    resultado = resultado * 2

  return resultado
```

```python
potencia_de_dos(3)
```

Reto 3: Evitando errores con try except

```python
import numpy as np
```

```python
def lista_strings_floats(lista_strings):
  lista_floats =[]

  for string in lista_strings:
    try:
      float_ = float(string)
      lista_floats.append(float_)
    except:
      lista_floats.append(np.nan)

  return lista_floats
```

```python
lista = ['1', '2', '3', '4', '5']
```

```python
lista_strings_floats(lista)
```

```python
lista2 = ['1.2', '33', '55.5', 'f', '78']
```

```python
lista_strings_floats(lista2)
```

```python
lista3 = ['f', 'g', 't', 'e', 'r']
```

```python
lista_strings_floats(lista3)
```

```python
def hacer_dataFrame(lista_de_Listas, indice):
  meses = ['enero', 'febrero', 'marzo', 'abril', 'mayo', 'junio', 'julio', 'agosto', 'septiembre', 'octubre', 'noviembre','diciembre']

  lista_de_series = []
  for i in range(len(lista_de_Listas)):
    lista_de_series.append(pd.Series(lista_de_Listas[i], index=indice, name=meses[i]))

  dataframe = pd.concat(lista_de_series, axis=1)
  return dataframe
```

```python
import requests
import pandas as pd
from datetime import datetime, timedelta

# Configura las fechas de inicio y fin y la clave de la API
start_date = '1995-01-01'
end_date = '1995-02-01'
api_key = 'SAFn7kDq0KkmXxI0QEmEczF3Y6Rqbf9hMpjseW9j'

# Convierte las fechas de inicio y fin a objetos datetime
start_date = datetime.strptime(start_date, '%Y-%m-%d')
end_date = datetime.strptime(end_date, '%Y-%m-%d')

# Lista para almacenar los DataFrames de cada intervalo
dfs = []

# Divide el rango de fechas en intervalos de 7 días
while start_date < end_date:
    # Calcula la fecha de fin del intervalo (7 días después)
    interval_end_date = start_date + timedelta(days=7)

    # Asegúrate de que la fecha de fin no supere la fecha final deseada
    if interval_end_date > end_date:
        interval_end_date = end_date

    # Construye la URL con los parámetros del intervalo actual
    url = f'https://api.nasa.gov/neo/rest/v1/feed?start_date={start_date.strftime("%Y-%m-%d")}&end_date={interval_end_date.strftime("%Y-%m-%d")}&api_key={api_key}'

    # Realiza la solicitud a la API
    response = requests.get(url)

    # Verifica si la solicitud fue exitosa (código de estado 200)
    if response.status_code == 200:
        # Convierte la respuesta JSON a un diccionario de Python
        data = response.json()

        # Accede a los datos específicos que te interesan
        neo_data = data['near_earth_objects']

        # Crea una lista de diccionarios para construir el DataFrame
        neo_list = []
        for date, objects in neo_data.items():
            for obj in objects:
                neo_list.append({
                    'Fecha': date,
                    'Nombre': obj['name'],
                    'Diámetro (metros)': obj['estimated_diameter']['meters']['estimated_diameter_max'],
                    'Distancia (kilómetros)': obj['close_approach_data'][0]['miss_distance']['kilometers'],
                })

        # Crea el DataFrame y añádelo a la lista
        df = pd.DataFrame(neo_list)
        dfs.append(df)

    else:
        print(f'Error en la solicitud. Código de estado: {response.status_code}')

    # Actualiza la fecha de inicio para el próximo intervalo
    start_date = interval_end_date

# Concatena todos los DataFrames en uno solo
result_df = pd.concat(dfs, ignore_index=True)

# Muestra el DataFrame resultante
print(result_df.head())
```
