Reto 2: Estimados de locación

```python
def promedio_custom(serie):
  sum = serie.sum()
  total_elementos = len(serie)

  promedio = sum/total_elementos
  return promedio
```

```python
promedio_custom(edades)
```

```python
import numpy as np
```

```python
def mediana_custom(serie):
  lista_orden = sorted(serie)
  total_elem = len(lista_orden)
  index = (int)(len(lista_orden) /2)

  if total_elem%2 == 0:
    return  (lista_orden[index - 1] + lista_orden[index]) /2
  else:
    return lista_orden[index]
```

```python
mediana_custom(edades)
```

Reto 3: Desviación estándar

```python
total =df_meteoritos.shape[0]
diameter_column = 'estimated_diameter.meters.estimated_diameter_max'
```

```python
mean = df_meteoritos[diameter_column].mean()
mean
```

```python
std = df_meteoritos[diameter_column].std()
std
```

Porcentaje de muestras que están a una distancia de 1 desviación estándar del promedio.

```python
#Categorizar datos de la columna que sean mayores o igual que promedio - desv.std --DATO MAS BAJO
within_one_std_filter_bottom = df_meteoritos[diameter_column] >= (mean - std)
#Categorizar datos de la columna que sean menores o igual que promedio +desv.std -DATO MAS ALTO
within_one_std_filter_top = df_meteoritos[diameter_column] <= (mean + std)
#Obtener conjunto de datos que esten en el intervalo de una desviacion estandar
within_one_std = df_meteoritos[within_one_std_filter_bottom & within_one_std_filter_top]
percentage_of_data_within_one_std = (within_one_std.shape[0] * 100 )/ total
percentage_of_data_within_one_std
```

Porcentaje de muestras que están a una distancia de 2 desviaciones estándares del promedio.

```python
#Categorizar datos de la columna que sean mayores o igual que promedio - desv.std --DATO MAS BAJO
within_two_std_filter_bottom = df_meteoritos[diameter_column] >= (mean - (std*2))
#Categorizar datos de la columna que sean menores o igual que promedio +desv.std -DATO MAS ALTO
within_two_std_filter_top = df_meteoritos[diameter_column] <= ((mean + std*2))
#Obtener conjunto de datos que esten en el intervalo de una desviacion estandar
within_two_std = df_meteoritos[within_two_std_filter_bottom & within_two_std_filter_top]
percentage_of_data_within_two_std = (within_two_std.shape[0] * 100 )/ total
percentage_of_data_within_two_std
```

Porcentaje de muestras que están a una distancia de 3 desviaciones estándares del promedio.

```python
#Categorizar datos de la columna que sean mayores o igual que promedio - desv.std --DATO MAS BAJO
within_three_std_filter_bottom = df_meteoritos[diameter_column] >= (mean - (std*3))
#Categorizar datos de la columna que sean menores o igual que promedio +desv.std -DATO MAS ALTO
within_three_std_filter_top = df_meteoritos[diameter_column] <= ((mean + std*3))
#Obtener conjunto de datos que esten en el intervalo de una desviacion estandar
within_three_std = df_meteoritos[within_three_std_filter_bottom & within_three_std_filter_top]
percentage_of_data_within_three_std = (within_three_std.shape[0] * 100 )/ total
percentage_of_data_within_three_std
```
