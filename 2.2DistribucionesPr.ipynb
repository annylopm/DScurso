```python
import pandas as pd
import seaborn as sns
import matplotlib as plt
from scipy import stats
```

Reto 1: Boxplots

```python
df1 = pd.read_csv('/content/drive/MyDrive/DataSets2-TCH/Remoto Datasets/Remoto near_earth_objects-jan_feb_1995-clean.csv', index_col=0)
df1.head()
```

```python
diameter_max = df1['estimated_diameter.meters.estimated_diameter_max']
sns.set(style='whitegrid')
sns.boxplot(x=(diameter_max))
```

```python
print('Promedio:', diameter_max.mean())
print('Mediana:', diameter_max.median())
print('Media truncada:', stats.trim_mean(diameter_max, 0.1))
print('Minimo:' , diameter_max.min())
print('Maximo:' , diameter_max.max())
print('percentil 25:' ,diameter_max.quantile(0.25))
print('percentil 75:' , diameter_max.quantile(0.75))
```

b) Eliminando valores atípicos

```python
#Grafica 1
iqr1 = df1['estimated_diameter.meters.estimated_diameter_max'].quantile(0.75) - df1['estimated_diameter.meters.estimated_diameter_max'].quantile(0.25)
filtro_inf1 =  df1['estimated_diameter.meters.estimated_diameter_max'] >  df1['estimated_diameter.meters.estimated_diameter_max'].quantile(0.25) - (1.5*iqr1)
filtro_sup1 = df1['estimated_diameter.meters.estimated_diameter_max'] <  df1['estimated_diameter.meters.estimated_diameter_max'].quantile(0.75) + (1.5*iqr1)

df1_filtrado = df1[filtro_inf1 & filtro_sup1]
```

```python
sns.boxplot(x=df1_filtrado['estimated_diameter.meters.estimated_diameter_max'])
```

```python
#Grafica2
iqr12 = df1['relative_velocity.kilometers_per_second'].quantile(0.75) - df1['relative_velocity.kilometers_per_second'].quantile(0.25)
filtro_inf12 = df1['relative_velocity.kilometers_per_second'] > df1['relative_velocity.kilometers_per_second'].quantile(0.25) - (1.5*iqr12)
filtro_sup12 = df1['relative_velocity.kilometers_per_second'] < df1['relative_velocity.kilometers_per_second'].quantile(0.75) + (1.5*iqr12)

df12_filtrado = df1[filtro_inf12 & filtro_sup12]
```

```python
sns.boxplot(x=df12_filtrado['relative_velocity.kilometers_per_second'])
```

```python
#grafica 3
iqr2 = df2['price.numberDouble'].quantile(0.75) - df2['price.numberDouble'].quantile(0.25)
filtro_inf2 = df2['price.numberDouble'] > df2['price.numberDouble'].quantile(0.25) - (1.5*iqr2)
filtro_sup2 = df2['price.numberDouble'] < df2['price.numberDouble'].quantile(0.75) + (1.5*iqr2)

df2_filtrado = df2[filtro_inf2 & filtro_sup2]
```

```python
sns.boxplot(x=df2_filtrado['price.numberDouble'])
```

```python
#grafica4
iqr3 = df3['land_size'].quantile(0.75) - df3['land_size'].quantile(0.25)
filtro_inf3 = df3['land_size'] > df3['land_size'].quantile(0.25) - (8*iqr3)
filtro_sup3 = df3['land_size'] > df3['land_size'].quantile(0.75) + (8*iqr3)

df3_filtrado = df3[filtro_inf3 & filtro_sup3]
```
