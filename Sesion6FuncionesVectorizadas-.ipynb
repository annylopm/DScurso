Reto 2: Agregaciones

```python
import pandas as pd
import numpy as np
```

```python
edades = pd.Series([23, 24, 23, 34, 30, 17, 18, 24, 35, 28, 27, 27, 34, 32, 29, 16, 16, 17, 19, 34, 45, 46, 43, 45, 43, 32, 25, 29, 28, 38, 30, 37, 38, 24, 26, 25, 24, 19, 19, 18, 17, 18, 21, 20, 23, 24, 25, 25, 26, 24, 23, 32, 24, 25, 24, 36, 35, 36, 38, 39, 45, 46, 43, 48, 42, 41, 41, 26, 19, 19, 19, 20, 39, 38, 43, 28, 27, 39, 43, 52, 50, 38, 15, 17, 23, 25, 19, 32, 34, 35, 19, 19, 20, 26, 25, 43, 45, 46, 34, 33, 30, 30, 34, 45, 50, 50, 47, 25, 34, 37, 38, 19, 19, 20, 25, 28, 34, 32, 36, 39, 39, 28, 34, 33, 22, 25, 17, 17, 22, 24, 25, 45, 46, 43, 34, 35, 32, 23])
```

```python
# obtener varianza
```

```python
#obtenemos la suma total de edades
suma_edades = np.sum(edades)
suma_edades
```

```python
#Contar el numero de edades del conjunto
muestra = pd.Series.count(edades)
muestra
```

```python
#obtener promedio de edades
promedio = suma_edades/muestra
promedio
```

```python
#obtener la diferencia entre cada edad y promedio
edades_dif = np.subtract(edades, promedio)
edades_dif
```

```python
#Elevar la serie al cuadrado
edades_cuadrado = np.power(edades_dif, 2)
edades_cuadrado
```

```python
#sumar los elementos de la serie y dividirlos entre la cantidad de datos de la serie original
suma = np.sum(edades_cuadrado)
division = suma/(muestra-1)
division
```

```python
#Raiz cuadrada del reusltado para obtener la varianza estandar
desv_estandar = np.sqrt(division)
desv_estandar
```

Reto 4: Identificando y limpiando NaNs

```python
import pandas as pd
import numpy as np

pd.options.mode.chained_assignment = None
```

```python
datos = {
    'precio': [12000, 5500, np.nan, 4800, 8900, np.nan, 1280, 1040, 23100, np.nan, 15000, 13400, np.nan],
    'cantidad_en_stock': [34, 54, np.nan, 78, 56, np.nan, 34, 4, 0, 18, 45, 23, 5],
    'cantidad_vendidos': [120, 34, np.nan, 9, 15, np.nan, 103, np.nan, np.nan, 23, 10, 62, 59],
    'descuentos': [np.nan] * 13
}
df = pd.DataFrame(datos, index=["Pokemaster", "Cegatron", "Pikame Mucho", "Lazarillo de Tormes", "Stevie Wonder", "Needle", "El AyMeDuele", "El Desretinador", "Sacamel Ojocles", "Desojado", "Maribel Buenas Noches", "Cíclope", "El Cuatro Ojos"])
```

```python
df
```

```python
df_copy = df.copy()
```

```python
#NaNs por columna
df_copy.isna().sum()
```

```python
#NaNs por fila
df_copy.isna().sum(axis=1)
```

```python
#eliminar filas donde todos los valores sean NaNs
df_copy = df_copy.dropna(axis = 0, how='all')
df_copy
```

```python
#eliminar columnas donde todos los valores sean NaNs
df_copy = df_copy.dropna(axis = 1, how = 'all')
df_copy
```

```python
#Dado que la columna cantidad_vendidos no es tan importante, cambia los NaNs que haya en esa columna por 0.
df_copy['cantidad_vendidos'] = df_copy['cantidad_vendidos'].fillna(0)
df_copy
```

```python
#Dado que la columna precio es muy importante, elimina las filas restantes que tengan algún NaN en dicha columna.
df_copy = df_copy.dropna()
df_copy
```
