# **Reto 1: Anotando gráficas**

```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
sns.set_style('white')
```

```python
df = pd.read_csv('/content/drive/MyDrive/DataSets2-TCH/Remoto Datasets/Remoto diabetes-clean.csv', index_col=0)
df
```

# a)

```python
glucosa = sns.distplot(df['glucose'], kde=False, norm_hist=False)
glucosa.set(title='Índice de glucosa en mujeres mayores de 21 años', xlabel='glucosa', ylabel='conteo')
```

## **b)**

```python
pregnancy_counts = df['pregnancies'].value_counts()
```

```python
pregnancy = sns.barplot(x=pregnancy_counts.index, y=pregnancy_counts)
pregnancy.set(title='Numero de embarazos en mujeres mayores a 21 años', xlabel='Num. embarazos', ylabel='Conteo de mujeres')
```

## **c)**

```python
healthy = df[df['outcome']==0]
sick = df[df['outcome']==1]
```

```python
ax = sns.distplot(healthy['glucose'], hist=False, kde_kws={'shade':True}, label='Saludable')
sns.distplot(sick['glucose'], hist=False, kde_kws={'shade':True}, ax=ax, label='Enferma')
ax.set_title('Grafica de densidades de indice sanos y no sanos en glucosa', fontsize =13, pad=15)
ax.legend(loc='upper right')
```

## **d)**

```python
graph = sns.distplot(sick['blood_pressure'], hist=False, kde_kws={'shade':True}, label='Blood preasure')
sns.distplot(sick['glucose'], hist=False, kde_kws={'shade':True}, label='Glucose')
sns.distplot(sick['insulin'], hist=False, kde_kws={'shade':True}, label = 'Insulin')
graph.set_title('Indice de factores que favorecen el desarrollo de diabetes')
graph.legend(loc='upper right')
```

# **Reto 2: Gráficas de barras**


```python
df2 = pd.read_csv('/content/drive/MyDrive/DataSets2-TCH/Remoto Datasets/Remoto melbourne_housing-clean.csv', index_col=0)
df2.head()
```

```python
counts = df2['rooms'].value_counts()
ax = sns.barplot(x=counts.index, y=counts, order=counts.index)
ax.set_title('Conteo de Cuartos por casas')
ax.set(xlabel='Rooms', ylabel='Frecuencia')
```

```python
count_type = df2['type'].value_counts()
count_type_percentages = count_type * 100/count_type.sum()
ax1 = sns.barplot(x=count_type_percentages.index, y=count_type_percentages)
ax1.set_title('Porcentaje de habitaciones segun su tipo')
ax1.set(xlabel='Tipos' , ylabel='Frecuencia en porcetaje')
```

```python
count_council_area = df2['council_area'].value_counts()
ax2 = sns.barplot(x=count_council_area, y=count_council_area.index, orient='h')
ax2.set(xlabel='Frecuencia por categoria')
```

# **Reto 3: Tablas de contingencia**

```python
df3 = pd.read_csv('/content/drive/MyDrive/DataSets2-TCH/Remoto Datasets/Remoto melbourne_housing-clean.csv', index_col=0)
df3.head()
```

```python
type_rooms = pd.crosstab(df3['type'], df3['rooms'])
type_rooms
```

```python
graph_h = sns.barplot(x=type_rooms.loc['h'].index, y=type_rooms.loc['h'])
graph_h.set_title('Frecuencia de cuartos en tipo de habitacion H')
graph_h.set(xlabel='Numero de cuartos', ylabel='Frecuencia')
```

```python
graph_t = sns.barplot(x=type_rooms.loc['t'].index, y=type_rooms.loc['t'])
graph_t.set_title('Frecuencia de cuartos en tipo de habitacion T')
graph_t.set(xlabel='Numero de cuartos', ylabel='Frecuencia')
```

```python
graph_u = sns.barplot(x=type_rooms.loc['u'].index, y=type_rooms.loc['u'])
graph_u.set_title('Frecuencia de cuartos en tipo de habitacion U')
graph_u.set(xlabel='Numero de cuartos', ylabel='Frecuencia')
```

## b) Segmentando por región, tipo y número de cuartos

```python
crosstab = pd.crosstab(df3['region_name'], [df3['type'], df3['rooms']])
crosstab
```

```python
crosstab_stack = crosstab.stack(level=0)
crosstab_stack
```

```python
fig, axes = plt.subplots(2,2)
```

```python
fig, axes = plt.subplots(2, 2, figsize=(10,8), sharex=True, sharey=True)

sns.barplot(x=crosstab_stack.loc[('Eastern Metropolitan', 'h')].index, y=crosstab_stack.loc[('Eastern Metropolitan', 'h')], ax=axes[0,0])
sns.barplot(x=crosstab_stack.loc[('Western Metropolitan', 'h')].index, y=crosstab_stack.loc[('Western Metropolitan', 'h')], ax=axes[0,1])
sns.barplot(x=crosstab_stack.loc[('Northern Metropolitan', 'h')].index, y=crosstab_stack.loc[('Northern Metropolitan', 'h')], ax=axes[1,0])
sns.barplot(x=crosstab_stack.loc[('Southern Metropolitan', 'h')].index, y=crosstab_stack.loc[('Southern Metropolitan', 'h')], ax=axes[1,1])

axes[0,0].set(xlabel='', ylabel='', title='Eastern Metropolitan')
axes[0,1].set(xlabel='', ylabel='', title='Western Metropolitan')
axes[1,0].set(xlabel='', ylabel='', title='Northern Metropolitan')
axes[1,1].set(xlabel='', ylabel='', title='Southern Metropolitan')
```

# **Reto 4: Graficando múltiples axis**
