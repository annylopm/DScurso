Ejemplo7: Estructuras de control de flujo

```python
var_1 = 4
var_2 = 3
```

```python
if var_1 < var_2:
  print("El bloque se ha corrido")
else:
  print("la condición no se ha cumplido")
```

```python
if var_1 > var_2:
  print("La primera condicion se ha cumplido")
elif var_1 < var_2:
  print("La segunda condicion se ha cumplido")
else:
  print("las primeras dos comparaciones fueron falsas")
```

```python
x = "hola"
match x:
  case "hola":
    print("saludo")
  case "adios":
    print("despedida")
  case _:
    print("otro mensaje")
```
